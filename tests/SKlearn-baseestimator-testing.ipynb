# Crossvalidation checking inclusion thresholds

```python
name = 'ecoli'  # 'bupa
nr = 7
```

```python
x_train, y_train, t_train = get_dataset(
    Path.cwd().parent / 'keel-data' / f'{name}-10-fold',
    f"{nr}tra",
    get_datatypes=True
)
x_test, y_test = get_dataset(
    Path.cwd().parent / 'keel-data' / f'{name}-10-fold',
    f"{nr}tst"
)
```

```python
classes = list(np.unique(np.append(y_train, y_test)))
y_train = np.array([classes.index(label) for label in y_train])
y_test = np.array([classes.index(label) for label in y_test])
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm

gs = GridSearchCV(
    estimator=RuleGenerator(),
    param_grid={'inclusion_threshold': [1-1e-6, 0.9999, 0.999,0.99, 0.95, 0.9]},
    scoring='balanced_accuracy',
    n_jobs=-1
)
```

```python
gs.fit(x_train, y_train)
```

```python
import pandas as pd

pd.DataFrame(gs.cv_results_)
```

```python
gs.best_params_
```

```python
model = RuleGenerator()
model.set_params(**gs.best_params_)
```

```python
model.get_params()
```

```python
model = RuleGenerator(
    approximation=LowerApproximation()
)
model.fit(x_train, y_train, _)
```

```python
balanced_accuracy_score(y_test, model.predict(x_test))
```
