# Generating rule length histograms

```python
from quickrules.data_handling import count_attributes
from pathlib import Path
from matplotlib import pyplot as plt
import numpy as np
```

Possible datasets:

```python
data_sets = [
    'australian',
     # 'automobile', # cat feats
     'bands',
     'bupa',
     'cleveland',
     # 'contraceptive', # no rules found
     # 'crx', # cat feats
     'dermatology',
     'ecoli',
     # 'german', # cat feats
     'glass',
     # 'haberman', # no rules found
     'heart',
     'ionosphere',
     # 'mammographic',  # no rules found
     'pima',
     # 'saheart', # cat feats
     'sonar',  # 70 features!
     'spectfheart',
     'vehicle',
     'vowel',
     'wine',
     'winequality-red',
     'wisconsin',
     'yeast'
]
```

Select the dataset of which we want to see the histogram.

```python
data_set = 'ecoli'
test_path = Path.cwd() / 'lower-approx-new-impl-test' / data_set
lengths = []
```

Count the total number of rule lengths across all folds, without averaging, as this would not change the shape of the histogram.

```python
for fold in range(10):
    lengths.extend(count_attributes(
                   test_path/ f'fold{fold+1}' / f'rules_fold{fold+1}.dat', '{'))
```

```python
data = lengths

# fixed bin size
bins = np.arange(0, max(data), 1) # fixed bin size

plt.xlim([0, max(data)])

plt.hist(data, bins=bins, alpha=0.5)
plt.title(f'Histogram of rule lengths in {data_set} dataset')
plt.xlabel('number of attributes')
plt.ylabel('count')
```

```python
plt.show()
```
