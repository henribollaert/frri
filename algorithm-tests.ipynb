```python
from quickrules.quickrules import QuickRules
from quickrules.relations import MinMaxRelation, RelationFactory, CategoricalRelation
from quickrules.fuzzy_operators import MinTNorm, KleeneDienesImplicator
from pathlib import Path
from quickrules.data_handling import get_dataset_dtypes
```

```python
fold = 1
x_train, y_train, t_train = get_dataset_dtypes(Path.cwd() / 'keel-data' / 'pima-10-fold', f'{fold}tra')
x_test, y_test, t_test = get_dataset_dtypes(Path.cwd() / 'keel-data' / 'pima-10-fold', f'{fold}tst')
```

```python
t_train
```

```python
rf = RelationFactory(MinMaxRelation, CategoricalRelation, MinTNorm())

model = QuickRules(
    implicator=KleeneDienesImplicator(),
    relation_factory=rf,
    relaxation=0.0,
    post_prune=False,
    rule_prune=True,
    verbose=True
)
```

```python
rf.get_relation(x_train, t_train)
```

```python
model.fit(x_train, y_train, t_train)
```

```python
model.get_rules_as_string()[-10:]
```

```python
preds = model.predict(x_test)
```

```python
from quickrules.data_handling import balanced_accuracy_score

balanced_accuracy_score(y_test, preds)
```

```python
model_no_rule_prune = QuickRules(
    implicator=KleeneDienesImplicator(),
    relation_factory=rf,
    relaxation=0.0,
    post_prune=False,
    rule_prune=False,
    verbose=True
)
```

```python
model_no_rule_prune.fit(x_train, y_train, t_train)
```

```python
no_preds = model_no_rule_prune.predict(x_test)
```

```python
balanced_accuracy_score(y_test, no_preds)
```
