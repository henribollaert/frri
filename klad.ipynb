```python
for d in data_sets:
    total = 0
    for fold in range (1, 11):
        _, y_test = get_dataset(Path.cwd() / 'keel-data' / f'{d}-10-fold', f'{fold}tst')
        numpy_pred = genfromtxt(
            Path.cwd() / 'close-min-max-combo-results' / f'{d}' / f'fold{fold}' / f'fold{fold}.dat',
            comments='@',
            missing_values='None',  # temp do not include!!!!!
            filling_values='0',
            # dtype='str'
        )
        total += accuracy_score(y_test, numpy_pred)
        # print(accuracy_score(y_test, numpy_pred))
    print(f'{d} average: {total/10}')
```

```python
example_x = np.array(
    [[1, 101, 50, 15, 36, 24.2, 0.526, 26],
     [8, 176, 90, 34, 300, 33.7, 0.467, 58],
     [7, 150, 66, 42, 342, 34.7, 0.718, 42],
     [7, 187, 68, 39, 304, 37.7, 0.254, 41],
     [0, 100, 88, 60, 110, 46.8, 0.962, 31],
     [0, 105, 64, 41, 142, 41.5, 0.173, 22],
     [1, 95, 66, 13, 38, 19.6, 0.334, 25]])

example_y = np.array([0, 1, 0, 1, 0, 0, 0])
```

```python
from quickrules.fuzzy_operators import MinTNorm, KleeneDienesImplicator
from quickrules.quickrules import QuickRules
from quickrules.relations import RelationFactory, StatisticalRelation
from quickrules.owaquickrules import OWAQuickRules
from quickrules.weights import LinearWeights, ExponentialWeights

minmax_factory = RelationFactory(StatisticalRelation, MinTNorm())
model = OWAQuickRules(MinTNorm(), KleeneDienesImplicator(), minmax_factory, ExponentialWeights())
```

```python
model.fit(example_x, example_y)
```

```python
x_train, y_train = get_dataset(Path.cwd() / 'keel-data' / 'cleveland-10-fold', '1tra')
```

```python
model.fit(x_train, y_train)
```

```python
x_test, y_test = get_dataset(Path.cwd() / 'keel-data' / 'cleveland-10-fold', '1tst')
```

```python
preds = model.predict(x_test)
```

```python
accuracy_score(y_true=y_test, y_pred=preds)
```
